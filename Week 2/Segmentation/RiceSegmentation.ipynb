# Libraries

```python
import numpy as np
from cv2 import imread
import matplotlib.pyplot as plt
from skimage.io import imread
```

# Read and display the image

```python
img = imread('onerice.bmp')
plt.imshow(img)
plt.show()
```

# How is the image?

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))

howis(img)
```

# Selection of first channel (RED)

```python
X = img
howis(X)
plt.imshow(X,cmap='gray')
plt.show()
```

# Thresholding - Simple Segmentation

```python
def segmenta(X,t):
  (N,M) = X.shape
  Y = np.zeros((N,M))
  area = 0
  for i in range(N):
    for j in range(M):
      if X[i,j] > t:
        Y[i,j] = 255
        area = area + 1
  print('area = ',area)
  return Y
```

# Segmentation

```python
Y = segmenta(X,120) # good threshold is 120
howis(Y)
plt.imshow(Y)
plt.show()
```

# Definition of a homogeneous background

Each row is the original row minus its minimum


```python
(N,M) = X.shape
Xm    = np.zeros((N,M),np.uint8)
for i in range(N):
  xmin = np.min(X[i,:])
  Xm[i,:] = X[i,:] - xmin
plt.imshow(Xm,cmap='gray')
plt.show()
```

```python
Y = segmenta(Xm,60)
plt.imshow(Y,cmap='gray')
plt.show()
```
