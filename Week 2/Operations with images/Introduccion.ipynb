## 2. Escala de grises

1. Cargar una imagen de prueba
--- 
- Vamos a cargar un gatito
- Y veremos su representacion como matriz

```python
from skimage.color import rgb2gray # rgb 2(to) gray, just that 

im = ski.data.cat() 
print(f'RGB image shape: {im.shape}')
im_2 = rgb2gray(im) 
print(f'Escala de grises shape: {im_2.shape}')

plt.imshow(im_2, cmap='gray') # muestra la imagen en escala de grises
plt.show() 
```

El RGB tiene 3 dimensiones y el escala de grises solo 1. Vamos a ver los pixeles

```python
im = ski.data.cat() 

print(f'RGB image shape: {im.shape}')
print(f'Pixel del rgb: {im[0,0]}') # pixel de la imagen rgb
#print(f'Primera fila de la imagen rgb: {im[0]}') # primera fila de la imagen rgb, 
print()
print(f'Escala de grises shape: {im_2.shape} <- 1 dimension')
print(f'Pixel de la escala de grises: {im_2[0,0]}') # pixel de la imagen en escala de grises

print ('\n\n observamos que una imagen 2d es una arreglo de 2 dimensiones + el numero de canales de color')

# ahora mostremos al gatito rgb
plt.imshow(im)
```

# Ejercicio 1. Dejale el ojo rojo al gato

```python
im = ski.data.cat() # por si ejecutamos la celda varias veces

# en y, desde 20 hasta 110, en x desde 10 hasta 110, en el canal 0 (rojo,green,blue), revisar arriba
# date cuenta que los ejes son 


# esto se llama slicing, y es una forma de seleccionar partes de un arreglo

# primero fijate como controlar el colr, luego el las dimensiones (guiate de los ejes)

im[90:150, 140:200, 0] = 255  #  <-------------------------------------- cual es el canal que controla el color?

# vamos ponerle negra la nariz al gato

im[220:270, 220:290]  = 0 # Como tiene que ser el pixel?

plt.imshow(im) # mostramos la imagen con el ojo rojo
```

## 2. Transformaciones
Invertir una imagen

```python
im = ski.data.cat()
im = rgb2gray(im) 
im = 255 - im  # vamos a restarle a cada pixel el valor de 255, que es el valor maximo de un pixel en escala de grises

plt.imshow(im, cmap='gray') 
```

## Binarzacion


```python
im = ski.data.coins()
print(f'Imagen de monedas shape: {im.shape}')
plt.imshow(im, cmap='gray')
```

Vamos a convertir esta imagen en blanco y negro, hay muchas tecnicas vamos a realizar una manual. El limite lo podemos seleccionar de forma manual(1) o podemos mandar un algoritmo para que encuentre un buen valor, por ejemplo Outsu (2)

```python
im = ski.data.coins()

threshold = 100

im = im > threshold # vamos a hacer una mascara, que es una imagen binaria 

# como escribo en una imagen 

plt.text(50, 20, 'im1', fontsize=12, color='red') 
plt.imshow(im, cmap='gray') 
```

```python
print(im) # la mascara es una imagen binaria, es decir, solo tiene 2 valores, 0 y 1
```

```python
im = ski.data.coins()
thresh = ski.filters.threshold_otsu(im)
print(f'El valor de la mascara es: {thresh}')
imbin = im > thresh
plt.text(50, 10, 'img2', color='red', fontsize=12)
plt.imshow(imbin, cmap='gray')

```
